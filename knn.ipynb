```python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.tree import DecisionTreeClassifier
```

```python
x = np.array([[3, 7], [9, 8], [3, 7], [2, 9], [3, 7]])
y = np.array([1, 0, 1, 0, 1])

df = pd.DataFrame(x, columns=['x1', 'x2'])
df = df.assign(y=y)

display(df)

# Buat Fungsi Distance

distance = lambda x1, x2: np.sqrt(np.sum((x1 - x2) ** 2))

df_distance = pd.DataFrame([[distance(x1, x2) for x2 in x] for x1 in x], columns=df.index, index=df.index)
df_distance = df_distance.assign(x1=df.x1)
df_distance = df_distance.assign(x2=df.x2)
df_distance = df_distance.assign(y=df.y)

display(df_distance)

# Buat Fungsi Menghitung KNN

knn = lambda x, y, x_test, k: np.bincount(y[np.argsort([distance(x_test, x_i) for x_i in x])[:k]]).argmax()


# Buat Fungsi Prediksi

prediksi = lambda x, y, x_test, k: np.array([knn(x, y, x_test_i, k) for x_test_i in x_test])

df['y_pred'] = prediksi(x, y, x, 3)


display(df)
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(x, y)
decision_tree.predict(x)
df['y_pred_dt'] = decision_tree.predict(x)
df
```
